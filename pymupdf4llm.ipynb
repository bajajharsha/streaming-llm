```python
import pymupdf4llm

md_text = pymupdf4llm.to_markdown("data/attention.pdf")
```

```python
# Now work with the markdown text, e.g. store as a UTF8-encoded file
import pathlib

pathlib.Path("output.md").write_bytes(md_text.encode())
print("Markdown saved to output.md")
```

```python
md_text_images = pymupdf4llm.to_markdown(
    doc="data/attention.pdf",
    page_chunks=True,
    write_images=True,
    image_path="images",
    image_format="png",
    dpi=300,
)
```

```python
md_text_chunks = pymupdf4llm.to_markdown(
    doc="data/attention.pdf", page_chunks=True
)
print(md_text_chunks[1])
```

```python
md_text_tables = pymupdf4llm.to_markdown(
    doc="data/attention.pdf"  # data with table
)
print(md_text_tables)
```
